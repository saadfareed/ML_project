# Data Scraping 

```python
# Getting the landing page after searching 'skin care' product.
url = "https://www.daraz.pk/catalog/?q=skin+care&_keyori=ss&from=input&spm=a2a0e.searchlist.search.go.3e7150f6yp46bT"
driver = webdriver.Chrome()
driver.get(url)
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
time.sleep(10)
driver.quit()
```

```python
df_daraz = pd.DataFrame(columns = ['Product', 'Review'])
# df_daraz.reset_index(inplace = True)
# Getting Each product one by one 
products = soup.find_all('div' , {'class' : 'c2prKC'})
# reviewList = list()
# i=0
for product in products:
#     i+=1
#     if i==4: 
#         print("BROKE!!!!!!!")
#         break
    # Extracting Name and link of product
    Name = product.find('div' , {'class': 'c16H9d'}).text
    link = product.find("a").get('href')
#     print('https:'+link)

#     Opening product's link for reviews retreival
    browser = webdriver.Chrome()
    browser.get('https:'+ link)
    time.sleep(12)
    
    print('\nProduct Name:',Name, '\nReviews:')
    
#     while loop is used to get all the reviews from pagination 
    while True:
        html = browser.page_source
        product_page = BeautifulSoup(html, 'html.parser')
        reviewsTag =  product_page.find_all('div', {'class':'item'})
    #     print('reviewtag' , reviewsTag)
    #     print(product_page.prettify())
    
#     'reviews' hold the HTML of review portion of the page
        reviews = reviewsTag if reviewsTag else "NA"
        if reviews == 'NA':
            print('\nNo review found')
            break
        
#       variable 'review' will iterate and get each review  
        for review in reviews:
            
            reviewText = review.find('div', {'class' : 'content'}).text 
    #         print(review)
    #         reviewList.append(reviewText)
            to_append = [Name, reviewText]
            print('=>', reviewText)
    #         print(to_append)
            
            
#             Stroing reviews into dataframe with there respective product

            df_length = len(df_daraz)
            df_daraz.loc[df_length] = to_append
        
#       getting Xpath of pagination 'Next' button and clicking it 

        try:
            if(browser.find_element_by_xpath('//*[@id="module_product_review"]/div/div[3]/div[2]/div/button[2]').is_enabled()):
    #             browser.find_element_by_xpath('//*[@id="module_product_review"]/div/div[3]/div[2]/div/button[2]').click()

                button = browser.find_element_by_xpath('//*[@id="module_product_review"]/div/div[3]/div[2]/div/button[2]')
                browser.execute_script("arguments[0].click();", button)
                time.sleep(5)
            else:
                break
        except:
            break
#     ////////////// while loop ends
     
    browser.quit()
```

```python
# Getting negative and positive words to perform sentiment analysis 
negativeWord = [line.strip() for line in open("negativeword.txt", 'r')]
positiveWord = [line.strip() for line in open("positiveword.txt", 'r')]

product = ''
reviewList = list()
reviewDict = {}
i=0

# making a dictionary where key is'product name' and value is 'prduct's reviews'
for index, row in df_daraz.iterrows():      
    if (i==0):
        product = row.Product
        i=+1
    
    if(product == row.Product):
        reviewList.append(row.Review)
    
    else:
        reviewDict[product] = reviewList
        reviewList = []
        reviewList.append(row.Review)
        product = row.Product
    
#     print(index,row)
        
    
    
```

```python
df_semantic = pd.DataFrame(columns = ['Product', 'Sentiment'])
index=0;

# Comparing Sentiment words with the reviews of each prdoucts then saving it into dataframe 
for product in reviewDict:
#     print(reviewDict[i])
    positiveCount = 0
    negativeCount = 0
    List =  reviewDict[product]
    for listIndex in range(len(List)):
        
        for j in range(len(positiveWord) - 1):      
            if positiveWord[j] in List[listIndex]:
#                 print(positiveWord[j])
                positiveCount += 1

        for k in range(len(negativeWord) - 1):     
            if negativeWord[k] in List[listIndex]:
#                 print(negativeWord[j])
                negativeCount += 1

    if positiveCount > negativeCount:
        df_semantic.at[index,'Product'] = product
        df_semantic.at[index,'Sentiment'] = 'Postive Reviews'      
    
    elif positiveCount < negativeCount:
        df_semantic.at[index,'Product'] = product
        df_semantic.at[index,'Sentiment'] = 'Negative Reviews'
        
    elif positiveCount == negativeCount:
        df_semantic.at[index,'Product'] = product
        df_semantic.at[index,'Sentiment'] = 'Equal Feedback'
    index += 1
```
